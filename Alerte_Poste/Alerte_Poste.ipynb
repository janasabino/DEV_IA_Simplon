```python
from collections import Counter
import re
```

```python
with open('connexion.log','r') as connexion, open('utilisateurs.txt','a') as utilisateurs:
      
    connexion_read = connexion.readlines()
    #pour récuperer seulement le nom de l'utilisateur
    users = [user.split(';')[1] for user in connexion_read]
    
    #pour récupérer la liste d'utilisateurs sans doublons
    users = set(users)
    
    users_list = "\n".join(str(user) for user in users)
      
    utilisateurs.write(users_list)
```

```python
users_log = open('connexion.log', 'r')
read_users_log = users_log.read()
single_users = read_users_log.split("\n")
```

**Regex**

**^:** asserts position at start of a line

**(.*):**
- . matches any character (except for line terminators)
- \* matches the previous token between zero and unlimited times, as many times as possible, giving back as needed (greedy)<\i><\u>

**\d{2}:**
- \d matches a digit (equivalent to [0-9])
- {2} matches the previous token exactly 2 times
    
**\s:** matches any whitespace character (equivalent to [\r\n\t\f\v ])

**(?:** [non-capturing groups](https://www.regular-expressions.info/brackets.html)

**HOURS:**

**0?[89]|1[0-8] (pour la partie heurs):**
- 1st Alternative 0?[89]
  - 0 matches the character 0 literally 
  - ? matches the previous token between zero and one times, as many times as possible, giving back as needed (greedy)
  - Match a single character present in the list below [89]
    - 89 matches a single character in the list 89 
    
- 2nd Alternative 1[0-8]
  - 1 matches the character 1 literally
  - Match a single character present in the list below [0-8]
    - 0-8 matches a single character in the range between 0 and 8

**[0-5][0-9] (pour la partie minutes):**
- Match a single character present in the list below [0-5] (premier chiffre des minutes)
  - 0-5 matches a single character in the range between 0 and 5
- Match a single character present in the list below [0-9] (deuxième chiffre des minutes)
  - 0-9 matches a single character in the range between 0 and 9

**$:** asserts position at the end of a line


```python
#identifier les personnes qui se connectent en dehors du horaire

users_regex = r'^(.*);(.*);\d{2}\/\d{2}\/\d{2}\s(?:(?:0?[89]|1[0-8]):[0-5][0-9]|19:00)$'

for u in single_users:
    if not re.search(users_regex, u):
        print(u)
```
